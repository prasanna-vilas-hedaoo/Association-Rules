**Apriori Algorithm **

```python
## apriori is function name ## min_support is a float value between 0 and 1 for minimun items support of the items returened
## the support is cimputed as
```

```python
freq_itm = apriori(book, min_support=0.1, use_colnames=True)

freq_itm
```

```python
freq_itm_1 = apriori(book, min_support=0.2, use_colnames=True)

freq_itm_1
```

```python
rules = association_rules(freq_itm, metric="lift", min_threshold=0.7)  ## min_threshold is confidence
rules
```

```python
rules.sort_values('lift', ascending = False)[0:20]
```

```python
rules[rules.lift>2]
```

```python
rules.sort_values('confidence', ascending = False)[0:20]
```
